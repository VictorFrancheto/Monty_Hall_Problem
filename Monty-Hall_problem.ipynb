# Simulation: The Monty Hall Problem

About the Problem: [Monty Hall Problem](https://en.wikipedia.org/wiki/Monty_Hall_problem)

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Set seed for reproducibility
random.seed(100)

# Problem parameters
doors = [1, 2, 3]
max_games = 2000  # Maximum number of simulations
step = 10         # Interval between simulation sizes
win_probabilities = []  # Stores the fraction of wins
game_counts = []         # Stores the number of games

# Monty Hall problem simulation
for num_games in range(1, max_games, step):
    wins = 0

    for _ in range(num_games):
        car_door = random.choice(doors)  # Place the car behind a random door
        initial_choice = random.choice(doors)  # Player chooses a random door

        # If the car is not behind the chosen door, switching wins the car
        if car_door != initial_choice:
            wins += 1

    # Calculate the win fraction and store results
    win_probabilities.append(wins / num_games)
    game_counts.append(num_games)
```

#### Plotting the Simulation

```python
# Results visualization
plt.figure(figsize=(12, 8))
plt.plot(game_counts, win_probabilities, 'go-', label='Simulation: Switch Door')
plt.axhline(y=2/3, color='blue', linestyle='--', label='Theoretical Probability (2/3)')
plt.ylim(0.5, 0.8) 
plt.title('Monty Hall Problem Simulation', fontsize=16)
plt.xlabel('Number of Games', fontsize=14)
plt.ylabel('Win Fraction', fontsize=14)
plt.legend(fontsize=12)
plt.show()
```
